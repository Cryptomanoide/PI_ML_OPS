```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Cargamos el dataset transformado y lo convertimos en un Data Frame de Pandas

```python
EDA_data = pd.read_csv(r'C:\Users\josgr\Desktop\PI_ML_OPS\data_transformada\eda_data.csv')
EDA_data
```

## Desanidamos las columnas que genres, productio_companies, production_countries y spoken_languages para disponibilizar de forma optima la informacion necesaria en sus registros

```python
import pandas as pd
import ast

# Desanidar y extraer el género de la película
EDA_data['genre'] = EDA_data['genres'].apply(lambda x: ast.literal_eval(x)[0]['name'] if pd.notnull(x) and ast.literal_eval(x) else None)

# Desanidar y extraer el nombre de la compañía
EDA_data['production_company'] = EDA_data['production_companies'].apply(lambda x: ast.literal_eval(x)[0]['name'] if pd.notnull(x) and ast.literal_eval(x) else None)

# Desanidar y extraer el país de producción
EDA_data['production_country'] = EDA_data['production_countries'].apply(lambda x: ast.literal_eval(x)[0]['name'] if pd.notnull(x) and ast.literal_eval(x) else None)

# Desanidar y extraer el idioma de la película
EDA_data['spoken_language'] = EDA_data['spoken_languages'].apply(lambda x: ast.literal_eval(x)[0]['name'] if pd.notnull(x) and ast.literal_eval(x) else None)
```

## Idiomas originales:

```python
language_counts = EDA_data['original_language'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=language_counts.index, y=language_counts.values)
plt.xlabel('Idioma original')
plt.ylabel('Cantidad')
plt.title('Idiomas originales de las películas (Top 10)')
plt.show()

```

## Popularidad de las películas:

```python
plt.hist(EDA_data['popularity'], bins=100)
plt.xlabel('Popularidad')
plt.ylabel('Frecuencia')
plt.title('Distribución de la popularidad de las películas')
plt.show()
```

## Duración de las películas:

```python
plt.hist(EDA_data['runtime'], bins=50)
plt.xlabel('Duración (minutos)')
plt.ylabel('Frecuencia')
plt.title('Distribución de la duración de las películas')
plt.show()
```

## Países de producción:

```python
import ast

# Convertir las cadenas a listas o diccionarios
EDA_data['production_countries'] = EDA_data['production_countries'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Filtrar y limpiar los datos
country_data = EDA_data['production_countries'].dropna().explode()
country_data = country_data.apply(lambda x: x['name'] if isinstance(x, dict) and 'name' in x else None)
country_data = country_data.dropna()

# Obtener el recuento de países de producción
country_counts = country_data.value_counts().head(10)

# Graficar el recuento de países
plt.figure(figsize=(20, 12))
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.xlabel('País de producción')
plt.ylabel('Cantidad')
plt.title('Países de producción de las películas (Top 10)')
plt.show()
```

## Relación entre puntuación y recuento de votos:

```python
# Crear el gráfico de dispersión
plt.scatter(EDA_data['vote_count'], EDA_data['vote_average'], c='b', alpha=0.5)

# Agregar etiquetas y título
plt.xlabel('Recuento de votos')
plt.ylabel('Puntuación promedio')
plt.title('Relación entre puntuación y recuento de votos')

# Configurar colores y transparencia
colors = ['r' if x > y else 'g' for x, y in zip(EDA_data['vote_count'], EDA_data['vote_average'])]
plt.scatter(EDA_data['vote_count'], EDA_data['vote_average'], c=colors, alpha=0.5)

# Mostrar el gráfico
plt.show()

```

## Años de lanzamiento:

```python
year_counts = EDA_data['release_year'].value_counts().sort_index()
plt.plot(year_counts.index, year_counts.values)
plt.xlabel('Año de lanzamiento')
plt.ylabel('Cantidad')
plt.title('Distribución de películas por año de lanzamiento')
plt.xticks(rotation=90)
plt.show()
```

## Ingresos y presupuestos:

```python
# Crear el gráfico de dispersión
plt.scatter(EDA_data['budget (billions)'], EDA_data['revenue (billions)'], c='b', alpha=0.5)

# Agregar etiquetas y título
plt.xlabel('Presupuesto (miles de millones)')
plt.ylabel('Ingresos (miles de millones)')
plt.title('Relación entre Presupuesto y Ingresos')

# Configurar colores y transparencia
colors = ['r' if x > y else 'g' for x, y in zip(EDA_data['budget (billions)'], EDA_data['revenue (billions)'])]
plt.scatter(EDA_data['budget (billions)'], EDA_data['revenue (billions)'], c=colors, alpha=0.5)

# Mostrar el gráfico
plt.show()

```

## Correlacion de las variables numericas 

```python
numeric_columns = ['popularity', 'runtime', 'vote_average', 'vote_count', 'budget (billions)', 'revenue (billions)']
correlation_matrix = EDA_data[numeric_columns].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlación entre variables numéricas')
plt.show()
```
